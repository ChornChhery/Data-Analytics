```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Expanded_data_with_more_features.csv')
df.head()
```

```python
# Drop unnamed column
df = df.drop("Unnamed: 0", axis=1)
df.head()
```

```python
# Change weekly study hours column
df["WklyStudyHours"] = df['WklyStudyHours'].str.replace("05-Oct","5-10")
df.head()
```

```python
# Gender Distribution
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df, x="Gender")
plt.title("Gender Distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# From the above chart we have analyzed that:
# the number of females in the data is more thab number of males

gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(6,7))
sns.heatmap(gb, annot=True)
plt.title("Relationship between Parents Education and Students Score")
plt.show()
```

```python
# From the above chart we have concluded that the 
# education of the parents  have a goog impact on their children
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(gb1, annot=True)
plt.show()
```

```python
# From the above chart we have concluded that there is no/negligible impact
# on the student's due to their parent's material status
```
